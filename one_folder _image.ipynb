```python
import os
import numpy as np
import shutil
from tqdm import tqdm
```

```python
# Путь к исходной папке
source_folder = 'D:/diploma/img'
# Путь к целевой папке
target_folder = 'D:/diploma/img_highres'

# Создаем целевую папку, если она еще не существует
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# Перебираем все подпапки в исходной папке
for subdir, dirs, files in os.walk(source_folder):
    # Перебираем все файлы в текущей подпапке
    for file in tqdm(files, desc=f"Processing {os.path.basename(subdir)}"):
        # Проверяем, является ли файл изображением (например,.jpg или.png)
        if file.endswith(('.jpg', '.jpeg', '.png')):
            # Формируем путь к исходному файлу
            source_file_path = os.path.join(subdir, file)
            # Формируем путь к целевому файлу, сохраняя структуру имен
            target_file_path = os.path.join(target_folder, os.path.basename(subdir) + '__' + file)
            # Копируем файл
            shutil.copy2(source_file_path, target_file_path)
```
